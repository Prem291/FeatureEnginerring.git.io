```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv(r'C:\Users\Priyanka\movies.csv',encoding='latin-1')
```

```python
df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col,pct_missing))
```

```python
df.dtypes
```

```python
df=df.dropna()
df['budget']= df['budget'].astype('int64')
df['gross']= df['gross'].astype('int64')
```

```python
df.head()
```

```python
df.columns
```

```python
df['yearcorrect']=df['released'].astype(str).str[:4]
```

```python
df.head()
```

```python
df.dtypes
```

```python
df=df.sort_values(by=['gross'],inplace=False, ascending=False)
```

```python
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df.head()
```

```python
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('budget')
```

```python
sns.regplot(x='budget',y='gross',data=df)
```

```python
df.corr(method='pearson')
```

```python
correaltion_matrix=df.corr(method='pearson')
sns.heatmap(correaltion_matrix,annot=True)
plt.title('Correlations for nNumericals')
plt.xlabel=('Movie features')
plt.ylabel=('Movie features')
```

```python
df_numerize=df
for col_name in df_numerize.columns:
    if(df_numerize[col_name].dtype=='object'):
        df_numerize[col_name]=df_numerize[col_name].astype('category')
        df_numerize[col_name]=df_numerize[col_name].cat.codes
df_numerize
```
